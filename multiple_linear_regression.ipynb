```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("multilinearregression.csv",sep = ";")
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['alan', 'odasayisi', 'binayasi']], df['fiyat'])
```

```python
yeni_veri = pd.DataFrame([[230, 4, 10]], columns=['alan', 'odasayisi', 'binayasi'])
tahmin = reg.predict(yeni_veri)

print("\nTahmini fiyat:", round(tahmin[0], 2), "TL")
```

```python
reg.coef_
reg.intercept_
# Multiple Linear regression formülü
# y= a + b1X1 + b2X2 + b3X3 + ...

a = reg.intercept_
b1 = reg.coef_[0]
b2 = reg.coef_[1]
b3 = reg.coef_[2]

x1 = 230
x2 = 4
x3 = 10
y = a + b1*x1 + b2*x2 + b3*x3

print("\n Formül ile hesaplanan fiyat:", round(y, 2), "TL")
```
